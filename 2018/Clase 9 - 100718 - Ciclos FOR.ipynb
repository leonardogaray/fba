# Ciclos FOR


## Repitiendo código 
El ciclo for es una de las formas que existen en la programación para poder repetir un conjunto de instrucciones tantas veces como querramos. Las mismas pueden ser ejecutadas un número fijo de veces, o dinámico. 
La instrucción FOR necesita 3 parámetros: el valor inicial, la condición para que termine, la cantidad de valores a incrementar. A continuación un conjunto de ejemplos:

* Repitiendo 5 veces, iniciando con 0, incrementando de a 1, hasta 5 (sin incluir)

```java
for(int i = 0; i < 5; i++){
    ellipse(200,200,50,50);//Esta línea se va a ejecutar 5 veces.
}
```

* Repitiendo 5 veces, iniciando con 1, incrementando de a 1, hasta 5 (incluído)

```java
for(int i = 1; i <= 5; i++){
    ellipse(200,200,50,50);//Esta línea se va a ejecutar 5 veces.
}
```

* Repitiendo 5 veces, iniciando con 5, decrementando de a 1, hasta 0 (sin incluir)

```java
for(int i = 5; i > 0; i--){
    ellipse(200,200,50,50);//Esta línea se va a ejecutar 5 veces.
}
```

* Repitiendo 5 veces, iniciando con 10, decrementando de a 2, hasta 0 (sin incluir)

```java
for(int i = 10; i > 0; i-=2){
    ellipse(200,200,50,50);//Esta línea se va a ejecutar 5 veces.
}
```

Todos los ejemplos anteriores se repiten 5 veces, pero como se puede apreciar, existen diferentes formas se hacerlo. La forma recomendada es la primera. Otra cosa a tener en cuenta, es que en todos los ejercicios, lo que se hace es dibujar una elipse en la misma posición 5 veces.

## Usando la variable del ciclo FOR para modificar el contenido
Ahora, haremos uso de la variable i (que puede tener el nombre que deseemos, mientras no sea igual al nombre de otra variable existente) para poder modificar la posición de las elipses que se dibujan

```java
void setup(){
    size(400,400);
    frameRate = 1;
}
void draw(){
    for(int i = 0; i < 5; i++){
        ellipse(i*50,200,50,50);//Esta linea se va a ejecutar 5 veces.
    }
}
```

Como el valor de i va a tener los valores [0,1,2,3,4], el resultado va a ser que la elipse pueda dibujarse en las cordenadas [0,50,100,150,200] en X, y 200 para Y.
Podemos hacer lo mismo para modificar los valores en Y, obteniendo los mismos valores que los mencionados anteriormente, pero para coordenadas en Y:

```java
void setup(){
    size(400,400);
}
void draw(){
    for(int i = 0; i < 5; i++){
        ellipse(200,i*50,50,50);//Esta linea se va a ejecutar 5 veces.
    }
}
```

Si aplicamos el mismo criterio en ambas coordenadas, las salidas serán [(0,0),(50,50),(100,100),(150,150),(200,200)], y visualmente se dibujarán sobre la diagonal principal de la pantalla (ya que es cuadrada).

```java
void setup(){
    size(400,400);
}
void draw(){
    for(int i = 0; i < 5; i++){
        ellipse(i*50,i*50,50,50);//Esta linea se va a ejecutar 5 veces.
    }
}
```

## <span style="color:red">!!Importante!!!</span>
El ciclo for, al hacerlo dentro del "void draw" va a ejecutarse todas las veces indicadas por cada ejecución del draw, o sea, que si tenemos un ciclo for que se ejecuta 5 veces, y el frameRate es de 60, en un segundo el contenido de ese ciclo for se va a ejecutar 300 veces.
Es importante diferenciar la ejecución del draw con el ciclo for.

## Eventos
Los eventos funcionan igual que lo visto hasta ahora, por ejemplo, si deseamos mover una columna de rectángulos con la misma posición del mouse, haríamos lo siguiente:

```java
void setup(){
    size(400,400);
}
void draw(){
    background(200);
    for(int i = 0; i < 5; i++){
        rect(mouseX, mouseY - i*50 - 50, 50, 50);
    }
}
```

## Ciclo FOR dinámico
Hasta ahora hemos visto siempre ciclos FOR que nos permiten una ejecución fija, pero podemos cambiar la misma a medida que se ejecuta el programa modificando la variable asociada al límite del ciclo. Por ejemplo, cada vez que se haga click con el mouse que se dibuje un cuadrado mas, y cada vez que se presiona una tecla, un cuadrado menos:

```java
int max = 1;
void setup(){
    size(400,400);
}
void draw(){
    background(200);
    for(int i = 0; i < max; i++){
        rect(mouseX, mouseY - i*50 - 50, 50, 50);
    }
}
void mousePressed(){
    max++;
}
void keyPressed(){
    max--;
}
```

## Ciclo FOR con Condicionales


```java
int max = 1;
void setup(){
    size(400,400);
}
void draw(){
    background(200);
    for(int i = 0; i < max; i++){
        fill(255);
        if(i%2==0){//Si i es par
            rect(mouseX, mouseY - i*50 - 50, 50, 50);
        }
        fill(0);
        text(i,mouseX + 20, mouseY - i*50 - 20);
    }
}
void mousePressed(){
    max++;
}
void keyPressed(){
    max--;
}
```
