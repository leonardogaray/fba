## Men√∫ con Drag

```java
int posX;
String texto;
void setup(){
    size(400,200);
    //Posicion de la zona que vamos a draguear
    posX = 100;
    texto = "Nada presionado";
}
void draw(){
    background(200);
    //Primer boton
    fill(0);
    text("A", 120, 130);
    noFill();
    rect(100,100,50,50);
    //Segundo boton
    fill(0);
    text("B", 170, 130);
    noFill();
    rect(150,100,50,50);
    //Tercer boton
    fill(0);
    text("C", 220, 130);
    noFill();
    rect(200,100,50,50);
    //Cuarto boton
    fill(0);
    text("D", 270, 130);
    noFill();
    rect(250,100,50,50);
    //Zona a draguear
    fill(255);
    rect(posX, 100, 200, 50);
    fill(0);
    //Texto indicador de boton presionado
    text(texto, 100, 50);
}
void mouseDragged(){
    if(mouseX > 100 && mouseX < 300){
        posX = mouseX;
    }
}
void mouseClicked(){
    if(mouseX > 100 && mouseX < posX){
        if(mouseX > 100 && mouseX < 150 && mouseY > 100 && mouseY < 150){
            texto = "Presionado A";
        }else if(mouseX > 150 && mouseX < 200 && mouseY > 100 && mouseY < 150){
            texto = "Presionado B";
        }else if(mouseX > 200 && mouseX < 250 && mouseY > 100 && mouseY < 150){
            texto = "Presionado C";
        }else if(mouseX > 250 && mouseX < 300 && mouseY > 100 && mouseY < 150){
            texto = "Presionado D";
        }else{
            texto = "Nada Presionado";
        }
    }
}
```

## Men√∫ Clonado

```java
int posX, posY, incrementoBoton1, incrementoBoton2, incrementoBoton3;
boolean iniciarAnimacionMostrar, finAnimacionMostrar, iniciarAnimacionOcultar;
String texto;
void setup(){
    size(400,200);
    posX = 100;
    posY = 100;
    iniciarAnimacionMostrar = false;
    iniciarAnimacionOcultar = false;
    incrementoBoton1 = 0; incrementoBoton2 = 0; incrementoBoton3 = 0;
    texto = "Nada Presionado";
}
void draw(){
    background(200);
    //Animacion mostrando los botones
    if(iniciarAnimacionMostrar && !finAnimacionMostrar){
       if(incrementoBoton1 < 100){ 
           incrementoBoton1+=1; 
       }
       if(incrementoBoton2 < 150){ 
           incrementoBoton2+=2; 
       }
       if(incrementoBoton3 < 200){ 
           incrementoBoton3+=3; 
       } 
       if(incrementoBoton3 >= 200){
           finAnimacionMostrar = true;
       }
    }
    //Animacion ocultando los botones
    if(iniciarAnimacionOcultar){
       if(incrementoBoton1 > 0){ 
           incrementoBoton1-=1; 
       }
       if(incrementoBoton2 > 0){ 
           incrementoBoton2-=2; 
       }
       if(incrementoBoton3 > 0){ 
           incrementoBoton3-=3; 
       } 
       if(incrementoBoton3 <= 0){
           //Reset de las ariables para volver a arrancar la animacion
           incrementoBoton1 = 0;
           incrementoBoton2 = 0;
           incrementoBoton3 = 0;
           iniciarAnimacionMostrar = false;
           iniciarAnimacionOcultar = false;
           finAnimacionMostrar = false;
       }
    }
    //Boton 1
    fill(255);
    ellipse(posX + incrementoBoton1, posY, 30, 30);
    fill(0);
    text("A", posX + incrementoBoton1 - 5, posY);
    //Boton 2
    fill(255);
    ellipse(posX + incrementoBoton2, posY, 30, 30);
    fill(0);
    text("B", posX + incrementoBoton2 - 5, posY);
    //Boton 3
    fill(255);
    ellipse(posX + incrementoBoton3, posY, 30, 30);
    fill(0);
    text("C", posX + incrementoBoton3 - 5, posY);
    //Menu
    fill(255);
    ellipse(posX, posY, 50, 50);
    //Texto indicador de boton presionado
    text(texto, 100, 50);
}
void mouseClicked(){
    if(dist(mouseX, mouseY, posX, posY) < 25){
        if(!iniciarAnimacionMostrar){
            iniciarAnimacionMostrar = !iniciarAnimacionMostrar;
        }else if(!iniciarAnimacionOcultar && finAnimacionMostrar){
            iniciarAnimacionOcultar = !iniciarAnimacionOcultar;
        }
    }
    if(finAnimacionMostrar){
        if(dist(mouseX, mouseY, posX + incrementoBoton1, posY) < 15){
            texto = "Presionado A";
        }else if(dist(mouseX, mouseY, posX + incrementoBoton2, posY) < 15){
            texto = "Presionado B";
        }else if(dist(mouseX, mouseY, posX + incrementoBoton3, posY) < 15){
            texto = "Presionado C";
        }else{
            texto = "Nada Presionado";
        }
    }
}
```
