# Arreglos - Battalla Naval

Armar el juego de la batalla naval con barcos de una sola celda para 2 Jugadores.

* 2 Jugadores
* 2 variables para las posiciones de los barcos
* 2 variables para las posiciones de las acciones del jugador
* Mostrar 2 tableros por Jugador. El tablero de la izquierda indica los barcos propios, y el de la derecha las acciones hechas con el contrario (agua, hundido)
* Tras hacer click en una celda, se tiene que habilitar un botón para pasarle el turno al Jugador 2

## Analizar el problema

Debemos identificar hacer que este problema complejo, lo podamos dividir en pequeños problemas, e ir resolviendo cada uno de ellos por partes. A continuación una forma posible:

1. Dibujar 2 tableros en la pantalla (Sin barcos)

A partir de este punto ... puedo tratar de analizar las soluciones para un solo jugador, y luego que las resuelva, extenderlo para ambos.
2. Definir como guardar las posiciones de los barcos del Jugador
3. Definir como guardar los lugares donde hizo click el Jugador
4. Mostrar en la grilla de la izquierda, los barcos del Jugador
5. Mostrar en la grilla de la derecha, los lugares donde hizo click el jugador.
6. Lograr identificar los click y marcar el valor dentro del arreglo

Con estos pasos resueltos, ya puedo avanzar en incorporar al segundo Jugador
7. Resolver las variables para guardar los datos del 2do Jugador
8. Permitir cambiar al Jugador siguiente


## Paso 1
### Dibujar 2 tableros en la pantalla (Sin barcos)

En este paso solo nos debemos concentrar en dibujar ambas grilas en la pantalla. Armemos 2 tableros de 8X8

```java
int cantFilas = 8;
int cantCols = cantFilas;
int ancho;
int alto;

void setup(){
    size(600,300);
    ancho = (width/2)/cantCols;//Obtenemos un ancho de acuerdo a la cantidad de columnas
    alto = height/cantFilas;//Obtenemos un alto de acuerdo a la cantidad de filas
}

void draw(){
    //Dibujamos la grilla de la izquierda
    for(int fila=0; fila < cantFilas; fila++){
        for(int col=0; col < cantCols; col++){
           rect(col*ancho, fila*alto, ancho, alto);
        }
    }
    //Dibujamos la grilla de la izquierda
    for(int fila=0; fila < cantFilas; fila++){
        for(int col=0; col < cantCols; col++){
           rect((width/2) + (col*ancho), fila*alto, ancho, alto);//Debemos dibujar la grilla desde la mitad de la pantalla
        }
    }
}
```

Vemos que se encuentran ambas grillas juntas, y no queda definido donde comienza uno y donde empieza otro. Retoquemos un poco para que ambas grillas queden diferencias, modificando el ancho que pueden usar, y el inico de cada una.
Vamos a tomar una regla de porcentajes, donde dejaremos un 5% del ancho de la pantalla para los márgenes, quedando nos un 80% para dibujar las grillas en lo ancho (0,05 + 0,4 + 0,05 + 0,05 + 0,4 + 0,05), y un 90% para el alto (0,05 + 0,9 + 0,05)

```java
int cantFilas = 8;
int cantCols = cantFilas;
int ancho, anchoCelda, margenAncho;
int alto, altoCelda, margenAlto;

void setup(){
    size(600,300);
    //Anchos
    margenAncho = round(width * 0.05); // De esta manera calculamos el 5% del ancho
    ancho = round((width/2) * 0.8);// En base a la mitad de la pantalla, calculamos el 80% del ancho
    anchoCelda = round(ancho/cantCols);//Obtenemos un ancho de acuerdo a la cantidad de columnas
    //Altos
    margenAlto = round(height * 0.05); // De esta manera calculamos el 5% del alto   
    alto = round(height * 0.9);// Calculamos el 90% del alto
    altoCelda = round(alto/cantFilas);//Obtenemos un alto de acuerdo a la cantidad de filas
}

void draw(){
    //Dibujamos la grilla de la izquierda
    for(int fila=0; fila < cantFilas; fila++){
        for(int col=0; col < cantCols; col++){
           rect(margenAncho + (col*anchoCelda), margenAlto + (fila*altoCelda), anchoCelda, altoCelda);
        }
    }
    //Dibujamos la grilla de la izquierda
    for(int fila=0; fila < cantFilas; fila++){
        for(int col=0; col < cantCols; col++){
           rect((width/2) + margenAncho + (col*anchoCelda), margenAlto + (fila*altoCelda), anchoCelda, altoCelda);//Debemos dibujar la grilla desde la mitad de la pantalla
        }
    }
}
```

## Paso 2
### Definir como guardar las posiciones de los barcos del Jugador

En este punto abordamos como guardar las posiciones de los barcos con arreglos. La grilla puede ser representado como un arreglo de arreglos, o arreglo bidimensional. Usaremos cada posición del arreglo para guardar los valores que nos indiquen los barcos, pero debemos tener presentes que valores debemos guardar.
Por ejemplo, si hicieramos un arreglo de booleanos, solo podríamos indicar si hay un barco, o no lo hay .... lo que nos deja afuera como indicar que el barco ha sido hundido. Para resolver ese problema debemos extender la cantidad de valores posibles.
Si definimos un arreglo de enteros, podemos hacer uso del 0 para indicar que no hya barco, el 1 para indicar que hay barco y el 2 para indicar que ha sido hundido.
La definición sería la siguiente:

```java
int [][] barcosJ1 = new int [cantFilas][cantCols];
```

Ahora debemos indicar la manera de poder agregar los barcos. Si lo hacemos de forma aleatoria, podemos incurrir en el problema de generar 2 o mas posiciones iguales, por lo que definamos directamente las posiciones de 5 barcos.

```java
barcosJ1[1][2] = 1;
barcosJ1[4][5] = 1;
barcosJ1[3][3] = 1;
barcosJ1[6][0] = 1;
barcosJ1[1][6] = 1;
```

## Paso 4
### Mostrar en la grilla de la izquierda, los barcos del Jugador

Debemos dibujar la grilla del lado izquierdo (Paso 1), y luego, si hay un barco, indicarlo de color negro, si esta hundido, de color rojo y si no hay nada, de color azul.

```java
int cantFilas = 8;
int cantCols = cantFilas;
int ancho, anchoCelda, margenAncho;
int alto, altoCelda, margenAlto;

//Barcos del Jugador1
int [][] barcosJ1 = new int [cantFilas][cantCols];

//Enemigos del Jugador 1
int [][] enemigosJ1 = new int [cantFilas][cantCols];

void setup(){
    size(600,300);
    //Anchos
    margenAncho = round(width * 0.05); // De esta manera calculamos el 5% del ancho
    ancho = round((width/2) * 0.8);// En base a la mitad de la pantalla, calculamos el 80% del ancho
    anchoCelda = round(ancho/cantCols);//Obtenemos un ancho de acuerdo a la cantidad de columnas
    //Altos
    margenAlto = round(height * 0.05); // De esta manera calculamos el 5% del alto   
    alto = round(height * 0.9);// Calculamos el 90% del alto
    altoCelda = round(alto/cantFilas);//Obtenemos un alto de acuerdo a la cantidad de filas
    
    //Inicializamos las posiciones de los barcos del Jugador1
    barcosJ1[1][2] = 1;
    barcosJ1[4][5] = 1;
    barcosJ1[3][3] = 1;
    barcosJ1[6][0] = 1;
    barcosJ1[1][6] = 1;
}

void draw(){
    //Dibujamos la grilla de la izquierda
    for(int fila=0; fila < cantFilas; fila++){
        for(int col=0; col < cantCols; col++){
            if(barcosJ1[fila][col]==1){//Hay un barco
                fill(0);
            }else if(barcosJ1[fila][col]==2){//Hay un barco hundido
                fill(255,0,0);
            }else{
                fill(0,0,255);//No hay barco
            }
            rect(margenAncho + (col*anchoCelda), margenAlto + (fila*altoCelda), anchoCelda, altoCelda);
        }
    }
    //Dibujamos la grilla de la izquierda
    for(int fila=0; fila < cantFilas; fila++){
        for(int col=0; col < cantCols; col++){
           rect((width/2) + margenAncho + (col*anchoCelda), margenAlto + (fila*altoCelda), anchoCelda, altoCelda);//Debemos dibujar la grilla desde la mitad de la pantalla
        }
    }
}
```

## Paso 6
### Lograr identificar los click y marcar el valor dentro del arreglo

Al hacer click, debemos identificar en que celda estamos haciendo click, y luego actuar en base a eso (lo cual aún no podremos hacer nada ya que no sabemos nada del otro jugador.
Haciendo uso de map, podemos facilitar la detección del click, pero a diferencia de ejercicios anteriores, la zona que debemos detectar no arranca en "0" y termina en "width", sino que arranca en:

(width/2) + margenAncho

y termina en:

width - margenAncho

Algo similar para el alto. Recordemos que intentames convertir la posición del mouse, a un valor entre 0 y 8 en la zona corresponidente a la grilla de la derecha.
Entonces:

```java
void mouseClicked(){
    int posCol = floor(map(mouseX, (width/2) + margenAncho, width - margenAncho, 0, cantCols));
    int posFila = floor(map(mouseY, margenAlto, height - margenAlto, 0, cantFilas));
}
```

```java
int cantFilas = 8;
int cantCols = cantFilas;
int ancho, anchoCelda, margenAncho;
int alto, altoCelda, margenAlto;

//Barcos del Jugador1 y 2
int [][] barcosJ1 = new int [cantFilas][cantCols];
int [][] barcosJ2 = new int [cantFilas][cantCols];
//Enemigos del Jugador 1 y 2
int [][] enemigosJ1 = new int [cantFilas][cantCols];
int [][] enemigosJ2 = new int [cantFilas][cantCols];

void setup(){
    size(600,300);
    //Anchos
    margenAncho = round(width * 0.05); // De esta manera calculamos el 5% del ancho
    ancho = round((width/2) * 0.8);// En base a la mitad de la pantalla, calculamos el 80% del ancho
    anchoCelda = round(ancho/cantCols);//Obtenemos un ancho de acuerdo a la cantidad de columnas
    //Altos
    margenAlto = round(height * 0.05); // De esta manera calculamos el 5% del alto   
    alto = round(height * 0.9);// Calculamos el 90% del alto
    altoCelda = round(alto/cantFilas);//Obtenemos un alto de acuerdo a la cantidad de filas
    
    //Inicializamos las posiciones de los barcos del Jugador1
    barcosJ1[1][2] = 1;
    barcosJ1[4][5] = 1;
    barcosJ1[3][3] = 1;
    barcosJ1[6][0] = 1;
    barcosJ1[1][6] = 1;
    //Inicializamos las posiciones de los barcos del Jugador2
    barcosJ2[3][1] = 1;
    barcosJ2[4][5] = 1;
    barcosJ2[5][3] = 1;
    barcosJ2[6][0] = 1;
    barcosJ2[2][6] = 1;
}

void draw(){
    //Dibujamos la grilla de la izquierda
    for(int fila=0; fila < cantFilas; fila++){
        for(int col=0; col < cantCols; col++){
            if(barcosJ1[fila][col]==1){//Hay un barco
                fill(0);
            }else if(barcosJ1[fila][col]==2){//Hay un barco hundido
                fill(255,0,0);
            }else{
                fill(0,0,255);//No hay barco
            }
            rect(margenAncho + (col*anchoCelda), margenAlto + (fila*altoCelda), anchoCelda, altoCelda);
        }
    }
    //Dibujamos la grilla de la izquierda
    for(int fila=0; fila < cantFilas; fila++){
        for(int col=0; col < cantCols; col++){
            if(enemigosJ1[fila][col]==1){//Hay un barco
                fill(0);
            }else if(enemigosJ1[fila][col]==2){//No hay barco
                fill(0,0,255);
            }else{
                fill(255);//No se hizo click aun
            }
            rect((width/2) + margenAncho + (col*anchoCelda), margenAlto + (fila*altoCelda), anchoCelda, altoCelda);//Debemos dibujar la grilla desde la mitad de la pantalla
        }
    }

}

void mouseClicked(){
    int posCol = floor(map(mouseX, (width/2) + margenAncho, width - margenAncho, 0, cantCols));
    int posFila = floor(map(mouseY, margenAlto, height - margenAlto, 0, cantFilas));
    
    //Si hay un barco enemigo
    if(barcosJ2[posFila][posCol]==1){
        barcosJ2[posFila][posCol] = 2; // Lo marco como hundido
        enemigosJ1[posFila][posCol] = 1; //Marco en mi grilla que esta hundido
    }else if(barcosJ2[posFila][posCol]==0){ //Si es agua
        enemigosJ1[posFila][posCol] = 2;
    }
}
```

```java
int cantFilas = 8;
int cantCols = cantFilas;
int ancho, anchoCelda, margenAncho;
int alto, altoCelda, margenAlto;
int turnoJugador = 1; //Arranca la variable para el jugador 1

//Barcos del Jugador1 y 2
int [][] barcosJ1 = new int [cantFilas][cantCols];
int [][] barcosJ2 = new int [cantFilas][cantCols];
//Enemigos del Jugador 1 y 2
int [][] enemigosJ1 = new int [cantFilas][cantCols];
int [][] enemigosJ2 = new int [cantFilas][cantCols];

void setup(){
    size(600,300);
    textSize(20);
    //Anchos
    margenAncho = round(width * 0.05); // De esta manera calculamos el 5% del ancho
    ancho = round((width/2) * 0.8);// En base a la mitad de la pantalla, calculamos el 80% del ancho
    anchoCelda = round(ancho/cantCols);//Obtenemos un ancho de acuerdo a la cantidad de columnas
    //Altos
    margenAlto = round(height * 0.05); // De esta manera calculamos el 5% del alto   
    alto = round(height * 0.9);// Calculamos el 90% del alto
    altoCelda = round(alto/cantFilas);//Obtenemos un alto de acuerdo a la cantidad de filas
    
    //Inicializamos las posiciones de los barcos del Jugador1
    barcosJ1[1][2] = 1;
    barcosJ1[4][5] = 1;
    barcosJ1[3][3] = 1;
    barcosJ1[6][0] = 1;
    barcosJ1[1][6] = 1;
    //Inicializamos las posiciones de los barcos del Jugador2
    barcosJ2[3][1] = 1;
    barcosJ2[4][5] = 1;
    barcosJ2[5][3] = 1;
    barcosJ2[6][0] = 1;
    barcosJ2[2][6] = 1;
}

void draw(){
    background(200);
    
    //Dibujamos la grilla de la izquierda
    for(int fila=0; fila < cantFilas; fila++){
        for(int col=0; col < cantCols; col++){
            if(barcosJ1[fila][col]==1){//Hay un barco
                fill(0);
            }else if(barcosJ1[fila][col]==2){//Hay un barco hundido
                fill(255,0,0);
            }else{
                fill(0,0,255);//No hay barco
            }
            rect(margenAncho + (col*anchoCelda), margenAlto + (fila*altoCelda), anchoCelda, altoCelda);
        }
    }
    //Dibujamos la grilla de la izquierda
    for(int fila=0; fila < cantFilas; fila++){
        for(int col=0; col < cantCols; col++){
            if(enemigosJ1[fila][col]==1){//Hay un barco
                fill(0);
            }else if(enemigosJ1[fila][col]==2){//No hay barco
                fill(0,0,255);
            }else{
                fill(255);//No se hizo click aun
            }
            rect((width/2) + margenAncho + (col*anchoCelda), margenAlto + (fila*altoCelda), anchoCelda, altoCelda);//Debemos dibujar la grilla desde la mitad de la pantalla
        }
    }
    
    //Dibujo el texto del Jugador activo
    fill(0);
    text("J:" + turnoJugador, (width/2) - (margenAncho/2), height/4);
    //Dibujo el boton de cambio de turno
    fill(200);
    rect((width/2) - (margenAncho/2), height/2, margenAncho, margenAlto);
}

void mouseClicked(){
    int posCol = floor(map(mouseX, (width/2) + margenAncho, width - margenAncho, 0, cantCols));
    int posFila = floor(map(mouseY, margenAlto, height - margenAlto, 0, cantFilas));
    
    //Si hay un barco enemigo
    if(barcosJ2[posFila][posCol]==1){
        barcosJ2[posFila][posCol] = 2; // Lo marco como hundido
        enemigosJ1[posFila][posCol] = 1; //Marco en mi grilla que esta hundido
    }else if(barcosJ2[posFila][posCol]==0){ //Si es agua
        enemigosJ1[posFila][posCol] = 2;
    }
    
    //Verifico si hago click dentro del boton
    if(mouseX>((width/2) - (margenAncho/2)) && mouseX < (((width/2) - (margenAncho/2)) + margenAncho) && mouseY > (height/2) && mouseY < ((height/2) + margenAlto)){
        //Hago el cambio de turno
        if(turnoJugador == 1){
            turnoJugador = 2;
        }else{
            turnoJugador = 1;
        }
    } 
}
```

### Dibujar el tablero del jugador de turno.

Esta es la parte mas complicada de comprender. Vamos a proceder a hacer un pequeño algoritmo simple para no tener que hacer dos veces las mismas cosas. Por ejemplo, si tengo que dibujar el tablero del jugador 1 o el jugador 2, en el void draw tendré un IF preguntando si es el jugador 1 o el 2, pero en ambos casos, el algoritmo para dibujar será el mismo, con la única diferencia que cambiaré las variables del jugador.
Para evitar esto, haremos uso de un tercer conjunto de variables que identifican al jugador de turno, las cuales se cargaran con los valores del jugador. En la declaración de variable agrego una nueva variable para guardar los datos de los barcos y los enemigos:

```java
//Barcos del Jugador1 y 2
int [][] barcosJ1 = new int [cantFilas][cantCols];
int [][] barcosJ2 = new int [cantFilas][cantCols];
//Enemigos del Jugador 1 y 2
int [][] enemigosJ1 = new int [cantFilas][cantCols];
int [][] enemigosJ2 = new int [cantFilas][cantCols];

//Barcos del Jugador de turno
int [][] barcosJT = new int [cantFilas][cantCols];
//Enemigos del Jugador de turno
int [][] enemigosJT = new int [cantFilas][cantCols];
//Barcos enemigos del Jugador de turno
int [][] barcosEnemigosJT = new int [cantFilas][cantCols];
```

Tras iniciar los valores de barcosJ1, copio el mismo a barcosJS

```java
//Inicializamos las posiciones de los barcos del Jugador1
barcosJ1[1][2] = 1;
barcosJ1[4][5] = 1;
barcosJ1[3][3] = 1;
barcosJ1[6][0] = 1;
barcosJ1[1][6] = 1;
//Inicializamos las posiciones de los barcos del Jugador2
barcosJ2[3][1] = 1;
barcosJ2[4][5] = 1;
barcosJ2[5][3] = 1;
barcosJ2[6][0] = 1;
barcosJ2[2][6] = 1;
//Copio barcosJ1 a barcosJT
barcosJT = barcosJ1;
```
