## Armar una grilla

Debemos encarar la solución haciendo uso de cliclos for anidados, que nos permitan dibujar rectángulosque puedan ser utilizados como celdas. 

```java
int cantCols = 10;
int ancho;

void setup(){
    size(300,300);
    ancho = width / cantCols;
}

void draw(){
    for(int col=0; col<cantCols;col++){
        rect(col*ancho,0,ancho,ancho);
    }
}
```

Con lo hecho anteriormente, vemos que se vá a diujar una fila con celdas, las cuales podemos mencionar como columnas. Si contenemos ese ciclo FOR dentro de otro, podremos repetir la creación de las filas tantas veces como querramos, y deberemos modificar la posición en Y de cada rect.
Con el código indicado a continuación, tenemos resuelto el primer pequeño problema.

```java
int cantCols = 10;
int cantFilas = 10;
int ancho;
int alto;

void setup(){
    size(300,300);
    ancho = width / cantCols;
    alto = height / cantFilas;
}

void draw(){
    for(int fila=0; fila<cantFilas; fila++){
        for(int col=0; col<cantCols;col++){
            rect(col*ancho, fila*alto,ancho,alto);
        }
    }
}
```

## Detectar cuando se hace click en una celda

Aca debemos recordar la función MAP, la cual nos permitía hacer conversión de escalas, y la podemos usar para poder convertir la posición en X del mouse, a un número en 0 y la contidad de columnas, y  repetir lo mismo para las filas.
De esta simple forma, cualquier posición del mouse en la pantalla va a ser convertida a un número entre 0 y 10 (cantCols y cantFilas).


```java
int cantCols = 10;
int cantFilas = 10;
int ancho;
int alto;
int col;
int fila;

void setup(){
    size(300,300);
    ancho = width / cantCols;
    alto = height / cantFilas;
}

void draw(){
    for(int fila=0; fila<cantFilas; fila++){
        for(int col=0; col<cantCols;col++){
            rect(col*ancho, fila*alto,ancho,alto);
        }
    }
}

void mouseClicked(){
    col = floor(map(mouseX,0,width,0,cantCols));
    fila = floor(map(mouseY,0,height,0,cantFilas));
}
```

Al obtener los valores de fila y columna, solo nos queda mostrar de alguna manera, que hemos hecho click en esa celda, y una manera puede ser cambiando el color de fondo de la celda en la que se ha hecho click.

```java
int cantCols = 10;
int cantFilas = 10;
int ancho;
int alto;
int colSeleccionada;
int filaSeleccionada;

void setup(){
    size(300,300);
    ancho = width / cantCols;
    alto = height / cantFilas;
}

void draw(){
    for(int fila=0; fila<cantFilas; fila++){
        for(int col=0; col<cantCols;col++){
            if(fila == filaSeleccionada && col == colSeleccionada){
                fill(0);
            }else{
                fill(255);
            }
            rect(col*ancho, fila*alto,ancho,alto);
        }
    }
}

void mouseClicked(){
    colSeleccionada = floor(map(mouseX,0,width,0,cantCols));
    filaSeleccionada = floor(map(mouseY,0,height,0,cantFilas));
}
```

## Identificar que al hacer click haya una mina

Asociar una mina puede ser hecho de una manera simple, como por ejemplo, creando 2 variables posX y posY que nos guarden una posición aleatoria, y si hacemos click justo en una celda con esas posiciones, damos por finalizada la partida.
Otra manera, sería la de hacer uso de una función matematica, que nos de cierta aleatoriedad, como por ejemplo, al hacer click en una celda con fila par, vamos a ver que hay una mina.

```java
int cantCols = 10;
int cantFilas = 10;
int ancho;
int alto;
int colSeleccionada;
int filaSeleccionada;
boolean perdiste = false;

void setup(){
    size(300,300);
    ancho = width / cantCols;
    alto = height / cantFilas;
}

void draw(){
    for(int fila=0; fila<cantFilas; fila++){
        for(int col=0; col<cantCols;col++){
            if(fila == filaSeleccionada && col == colSeleccionada){
                fill(0);
            }else{
                fill(255);
            }
            rect(col*ancho, fila*alto,ancho,alto);
        }
    }
}

void mouseClicked(){
    colSeleccionada = floor(map(mouseX,0,width,0,cantCols));
    filaSeleccionada = floor(map(mouseY,0,height,0,cantFilas));
    
    if(filaSeleccionada%2 == 0){
        perdiste = true;
    }
}
```

Otra manera, podría ser definir una número aleatorio al principio entre 1 y 4, y asignar ese valor al módulo. Tener en cuenta que esto lo hacemos ya que no tenemos los recursos como "arreglos" para obtener una solución mas límpia.

```java
int cantCols = 10;
int cantFilas = 10;
int ancho;
int alto;
int colSeleccionada;
int filaSeleccionada;
boolean perdiste = false;
int modulo = floor(random(1,5));

void setup(){
    size(300,300);
    ancho = width / cantCols;
    alto = height / cantFilas;
}

void draw(){
    for(int fila=0; fila<cantFilas; fila++){
        for(int col=0; col<cantCols;col++){
            if(fila == filaSeleccionada && col == colSeleccionada){
                fill(0);
            }else{
                fill(255);
            }
            rect(col*ancho, fila*alto,ancho,alto);
        }
    }
}

void mouseClicked(){
    colSeleccionada = floor(map(mouseX,0,width,0,cantCols));
    filaSeleccionada = floor(map(mouseY,0,height,0,cantFilas));
    
    if((filaSeleccionada + colSeleccionada ) % modulo == 0){
        perdiste = true;
    }
}
```
