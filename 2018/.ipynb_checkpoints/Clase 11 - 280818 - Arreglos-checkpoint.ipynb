# Arreglos

## Variables de mas de un valor

Hasta ahora, las variables que venimos utilizando siempre nos permiten guardar un único valor al mismo tiempo. Esto cubre las necesidades de los problemas que hemos resuelto hasta el momento, pero en algunos casos, no alcanza.
Supongamos que deseamos guardar la posición de un personaje en la pantalla, para lo cual hacemos uso de 2 variables enteras, las cuales son <b>posX</b> y <b>posY</b>. Bien. Problema resuleto.
Ahora, supongamos que necsitamos guardar la posición de 20 personajes .... tendríamos que iniciar con las variable <b>posX1</b> y <b>posY1</b>, <b>posX2</b> y <b>posY2</b>, <b>posX3</b> y <b>posY3</b>, ..... y así sucesivamente.
Para estos casos, tenemos los arreglos, que nos permiten guardar muchos valors, pero haciendo uso a una única variable

## Tipos de Datos y Declaración

Los arreglos, como las variables, permiten guardar datos de algún tipo, pero siempre del mismo. Los arreglos pueden ser de enteros (int), cadenas de characteres (String), flotantes (float), etc.. Es mas, un arreglo puede ser del tipo arreglo.
A continuación ejemplos en la forma de declarar un arreglo:

```java
int [] posXs = new int [10]; // Un arreglo de enteros de 10 elementos
String [] nombres = new String [5]; // Un arreglo de Strings de 5 elementos
float [] decimales = new float [100]; // Un arreglo de decimales de 100 elementos
```

## Asignación

Para poder realizar una asignación de valores, es necesario definir en que posición del arreglo vamos a guardar dicho valor. Tener en cuenta que los arreglos comienzan en la posición 0 y finalizan en la posición igual a la longitud definida menos 1.
Un arreglo de 50 elementos comenzará en la posición 0 y finalizará en la 49. Si tratamos de poner una dirección mayor a 49 o menor a 0, Processing nos dirá que no es posible acceder.

```java
posXs[0] = 10; // Se asigna el valor 10 a la primer posicion del arreglo
nombres[0] = "Juan"; // Se asigna el valor "Juan" a la primer posicion del arreglo
decimales[99] = 23.334; // Se asigna el valor 23.334 a la ultima posición del arreglo
```

También es posible realizar una asignación múltiple, haciendo uso de un ciclo <b>for</b>

```java
for(int i = 0; i < 10; i++){
    posXs[i] = i; //En este arreglo, estamos asignando dentro de la posición i, el valor i
}

for(int i = 0; i < 5; i++){
    nombres[i] = "Juan " + i; //En este arreglo, estamos asignando dentro de la posición i, el valor "Juan 1", "Juan 2", "....
}

for(int i = 0; i < 100; i++){
    decimales[i] = random(1,100); //En este arreglo, estamos asignando un valor aleatorio entre 1 y 100
}
```

## Lectura

Se mantienen las mismas reglas con respecto a las dimensiones, pero en este caso en lugar de asignar un valor, tomaremos el mismo que se ecuentra en una posición definida.

```java
int miEntero = posXs[0]; // Se lee el valor de la primer posicion del arreglo
String miNombre = nombres[0]; // Se lee el valor de la primer posicion del arreglo
float miMedida = decimales[99]; // Se lee el valor de la ultima posicion del arreglo
```

También es posible realizar una lectura múltiple, haciendo uso de un ciclo <b>for</b>

```java
for(int i = 0; i < 10; i++){
    text(posXs[i], 50, i*10);
}

for(int i = 0; i < 5; i++){
    text(nombres[i], 150, i*10);
}

for(int i = 0; i < 100; i++){
    text(decimales[i], 250, i*10);
}
```

## Ahora todo junto

```java
int [] posXs = new int [10];
String [] nombres = new String [5];
float [] decimales = new float [100];

void setup(){
    size(400,400);
    
    for(int i = 0; i < 11; i++){
        posXs[i] = i;
    }

    for(int i = 0; i < 5; i++){
        nombres[i] = "Juan " + i;
    }

    for(int i = 0; i < 100; i++){
        decimales[i] = random(1,100);
    }
}

void draw(){
    for(int i = 0; i < 10; i++){
        text(posXs[i], 50, i*10);
    }

    for(int i = 0; i < 5; i++){
        text(nombres[i], 150, i*10);
    }

    for(int i = 0; i < 100; i++){
        text(decimales[i], 250, i*10);
    }
}
```

## Trabajando con eventos 

Armemos un programa que permita mostrar en la pantalla cinco ceros en fila. Al presionar un número del teclado entre 0 y 4, se debe incrementar en uno el valor que se encuentre en esa posición

```java
int cantTeclas = 5;
int [] teclas = new int [cantTeclas];

void setup(){
    size(400,400);
    textSize(30);
    //Cargamos el arreglo con ceros
    for(int i = 0; i < cantTeclas; i++){
        teclas[i] = 0;
    }
}

void draw(){
    background(200);
    //Imprimimos en pantalla los valores que posee el arreglo
    for(int i = 0; i < cantTeclas; i++){
        text(teclas[i], 50*i, 100);
    }
}

void keyPressed(){
    //Al presionar una tecla, la comparamos con un número, y si corresponde
    //incrementamos en uno el valor que se encuentra en la misma posicion
    //que el numero presionado
    if(key == '0'){
        teclas[0]+=1;
    }
    if(key == '1'){
        teclas[1]+=1;
    }
    if(key == '2'){
        teclas[2]+=1;
    }
    if(key == '3'){
        teclas[3]+=1;
    }
    if(key == '4'){
        teclas[4]+=1;
    }
}
```

El programa anterior tiene algo que nos llama la atención dentro del evento <b>keyDown</b>, y es que tenemos tantos <b>if</b> como número presionados. 
Para poder resolver esto recurrimos a un <b>for</b> que nos ayude, haciendo lo siguiente:

```java
for(int i = 0; i < cantTeclas; i++){
   if(key == i){
        teclas[i]+=1;
    } 
}
```

Pero esto no va a funcionar, ya que <b>key</b> retora un caracter e <b>i</b> es un entero, por lo que si presionamos la tecla <b>'0'</b> se realizará la comparación <b>'0' == 0</b>. Para poder resolver el problema, debemos entender un poco que es el código ASCII.

Cada tecla que posee el teclado tiene asignado un código que se encuentra definido en una tabla, denominada ASCII. La tabla (Wikipedia - https://es.wikipedia.org/wiki/ASCII) indicada a continuación:

Podemos ver que la tecla <b>0</b> posee el decimal 48, la tecla <b>1</b> el 49 y así sucesivamente.

Si convertimos la instrucción <b>key</b> a un entero, obtendremos estos valores, pero, tendremos que restarle 48 para poder comparar con el decimal esperado. Así <b>48 - 48 == 0</b>, <b>49 - 48 == 1</b>, etc.

```java
for(int i = 0; i < cantTeclas; i++){
   if((int(key)  48) == i){
        teclas[i]+=1;
    } 
}
```

```java
int cantTeclas = 5;
int [] teclas = new int [cantTeclas];

void setup(){
    size(400,400);
    textSize(30);
    for(int i = 0; i < cantTeclas; i++){
        teclas[i] = 0;
    }
}

void draw(){
    background(200);
    for(int i = 0; i < cantTeclas; i++){
        text(teclas[i], 50*i, 100);
    }
    text("Key: ", 100, 300);
    text(key, 200, 300);
    text("Codigo: " + int(key), 100, 350);
}

void keyPressed(){
    for(int i = 0; i < cantTeclas; i++){
       if((int(key) - 48) == i){
            teclas[i]+=1;
        } 
    }
}
```

## Volvemos al buscaminas

```java
int cantCols = 10;
int cantFilas = 10;
int ancho;
int alto;
int colSeleccionada;
int filaSeleccionada;
boolean perdiste = false;
int [][] minas = new int [cantCols][cantFilas];

void setup(){
    size(300,300);
    textSize(30);
    ancho = width / cantCols;
    alto = height / cantFilas;
    
    for(int i = 0; i < 10; i++){
        for(int j = 0; j < 10; j++){
            if(random(1,100) > 50){
                minas[i][j] = 1;
            }else{
                minas[i][j] = 0;
            }
        }
    }
}

void draw(){
    if(!perdiste){
        for(int fila=0; fila<cantFilas; fila++){
            for(int col=0; col<cantCols;col++){
                fill(255);
                rect(col*ancho, fila*alto,ancho,alto);
            }
        }
    }else{
        for(int fila=0; fila<cantFilas; fila++){
            for(int col=0; col<cantCols;col++){
                if(minas[fila][col] == 1){
                    fill(0);
                }else{
                    fill(255);
                }
                rect(col*ancho, fila*alto,ancho,alto);
            }
        }
        
        fill(255,0,0);
        text("FINAL", 100, 200);     
    }
}

void mouseClicked(){
    colSeleccionada = floor(map(mouseX,0,width,0,cantCols));
    filaSeleccionada = floor(map(mouseY,0,height,0,cantFilas));
    
    for(int fila=0; fila<cantFilas; fila++){
        for(int col=0; col<cantCols;col++){
            if(minas[filaSeleccionada][colSeleccionada] == 1){ 
                perdiste=true;
            }
        }
    }
}
```
