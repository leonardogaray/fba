## Cambiando el color de relleno - Parte 1

Dividimos la pantalla en 2 mitades y de un lado dibujamos con un color, y del otro lado, con otro color.

```java
void setup(){
    size(400,400);
}

void draw(){
    // Dibujamos una linea marcando el centro de la pantalla
    line(width/2, 0, width/2, height);
    // Dibujamos la elipse que acompana al mouse
    ellipse(mouseX, mouseY, 50, 50);
    // Si el mouse pasa la mitad de la pantalla, cambiamos el fill
    if(mouseX > width/2){
        fill(200);
    }else{
        fill(0);
    }
}
```

## Cambiando el color de relleno - Parte 2

Dividimos la pantalla en 4 partes y dibujamos con colores distintos en cada mitad

```java
void setup(){
    size(400,400);
}

void draw(){
    // Dibujamos una linea marcando el centro vertical de la pantalla
    line(width/2, 0, width/2, height);
    // Dibujamos una linea marcando el centro horizontal de la pantalla
    line(0, height/2, width, height/2);
    // Dibujamos la elipse que acompana al mouse
    ellipse(mouseX, mouseY, 50, 50);
    // Si el mouse pasa la mitad de la pantalla, cambiamos el fill
    if(mouseX < width/2 && mouseY < height/2){
        fill(255,0,0);
    }else if(mouseX > width/2 && mouseY < height/2){
        fill(0,255,0);
    }else if(mouseX < width/2 && mouseY > height/2){
        fill(0,0,255);
    }else{
        fill(0);
    }
}
```

## Cambiando el color de relleno en una zona cuadrada

Al entrar en una zona de 200 de ancho y 200 de alto, dibujamos con otro color

```java
void setup(){
    size(400,400);
}

void draw(){
    // Dibujamos una cuadrado marcando la zona
    noFill();
    rect(100, 100, 200, 200);
    // Si el mouse pasa la mitad de la pantalla, cambiamos el fill
    if(mouseX > 100 && mouseX < 300 && mouseY > 100 && mouseY < 300){
        fill(200);
    }else{
        fill(0);
    }
    // Dibujamos la elipse que acompana al mouse
    ellipse(mouseX, mouseY, 50, 50);
    
}
```

## Cambiando el color de relleno en una zona circular

Al entrar en una zona centrada de 200 de diametro, dibujamos con otro color

```java
void setup(){
    size(400,400);
}

void draw(){
    // Dibujamos una circular marcando la zona
    noFill();
    ellipse(width/2, height/2, 200, 200);
    // Si el mouse pasa la mitad de la pantalla, cambiamos el fill
    if(dist(mouseX,mouseY,width/2, height/2)<100){
        fill(200);
    }else{
        fill(0);
    }
    // Dibujamos la elipse que acompana al mouse
    ellipse(mouseX, mouseY, 50, 50);
    
}
```

## Arrastramos una figura con el mouse

Al arrastrar una figura debemos tener en cuenta 3 cosas

* Tener guardada la posicin de la figura
* Poder detectar cuando estamos dentro
* Modificar la posicin de la figura de acuerdo a donde se encuentr el mouse 

```java
int posX, posY, diametro;
void setup(){
    size(400,400);
    posX = 200;
    posY = 200;
    diametro = 100;
}
void draw(){
    background(200);
    ellipse(posX, posY, diametro, diametro);
}
void mouseDragged(){
   if(dist(mouseX, mouseY, posX, posY)<diametro/2){
       posX = mouseX;
       posY = mouseY;
   } 
}
```

## Moviendo una figura con las flechas del teclado

Podemos hacer uso de las flechas del teclado haciendo uso de keyCode y compararlo con las constantes LEFT y RIGHT (UP y DOWN en caso de querer subir y bajar). Para identificar cuando se presiona una tecla, se usa la variable global keyPressed

```java
int posX, dir;
void setup(){
    size(400,400);
    posX=200;
    dir=1;
}
void draw(){
    // Detectamos si se presiona una tecla
    if(keyPressed){
        // Nos fijamos si la tecla esta dentro del grupo de teclas especiales
        if (key == CODED){
            // Si la tecla presionada es hacia la derecha
            if(keyCode == RIGHT){
                dir = 1;
            // Si la tecla presionada es hacia la izquierda
            }else if(keyCode == LEFT){
                dir = -1;
            }
            // Verificamos si alguna de las teclas se ha presionado para hacer el cambio correspondiente
            if(keyCode == RIGHT || keyCode == LEFT){
                posX = posX + 5 * dir;
            }
        }
    }
    ellipse(posX, 200, 50, 50);
}
```

## Mover un escenario con las teclas haciendo uso de la tcnica parallax (gracias por la definicin al alumno sin nombre)

En los videos juegos en 2D que poseen un personaje que va corriendo, se pueden ver unos escenarios que dan la sensacin de profundidad al irse moviendo en diferentes velocidades. Esta tcnica, llamada parallax, define los siguientes puntos:

* El escenario pdebe dividirse en diferentes capas.
* Una de las capas, debe ser por la cual se desplaza el personaje principal y debe moverse a la velocidad que uno desee que se mueva el personaje.
* Las capas que estn por detras deben ir moviendose mas despacio que la siguiente y mas rpido que la anterior.
* Las capas que estn por delante, deben ir moviendose mas rpido que la anterior y mas despacio que la siguiente.

Dicho de otra manera, si poseemos 5 capas de escenario, y en la capa 3 est nuestro personaje, podemos decir que la capa 3 se mueve a la velocidad X, la capa 2 a 1/2X y la capa 1 a 1/4 X, mientras que la capa 4 se mueve a 1.5X y la 5 a 2X.

Al aplicar estos conceptos, podemos ver el efecto que se muestra en el ejemplo siguiente

```java
PImage parallax1, parallax2, parallax3, parallax4;
Float posX0, posX1, posX2, posX3, posX4;
Integer dir = 1;

void setup(){
    size(400, 300);
    // Cargamos todas las imagenes
    parallax1 = loadImage("clase6/parallax-1.png");
    parallax2 = loadImage("clase6/parallax-2.png");
    parallax3 = loadImage("clase6/parallax-3.png");
    parallax4 = loadImage("clase6/parallax-4.png");
}
void draw(){
    background(200);
    
    //Ubicamos las imagenes en la posicion que le corresponde a cada una
    image(parallax1, posX1, 0);
    image(parallax2, posX2, 20);
    image(parallax3, posX3, 40);
    image(parallax4, posX4, 230);
    
    // Detectamos si se presiona una tecla
    if(keyPressed){
        // Nos fijamos si la tecla esta dentro del grupo de teclas especiales
        if (key == CODED){
            // Si la tecla presionada es hacia la derecha
            if(keyCode == RIGHT){
                dir = -1;
            // Si la tecla presionada es hacia la izquierda
            }else if(keyCode == LEFT){
                dir = 1;
            }
            // Verificamos si alguna de las teclas se ha presionado para hacer el cambio correspondiente
            if(keyCode == RIGHT || keyCode == LEFT){
                posX3 = posX3 + 5 * dir;
                posX2 = posX2 + 3 * dir;
                posX1 = posX1 + 1 * dir;
                posX4 = posX4 + 10 * dir;
            }
            
        }
    }
}
```
