### Figuras en diferentes direcciones

```java
int posX, posY;
void setup(){
    size(400,300);
    background(200);
    posX=0; posY=0;
}
void draw(){
    ellipse(posX,50, 50, 50);
    rect(width-100, posY, 50, 50);
    ellipse(50, height-posY,50, 50);
    rect(width-posX, height-100, 50, 50);
    posX++;
    posY++;
}
```

### Líneas aleatorias

```java
int posX, posY;
void setup(){
    size(300,300);
}
void draw(){
    stroke(random(0,255),random(0,255),random(0,255));
    posY = round(random(0,height));
    line(0,posY, width, posY);
}
```

### Tamaño variable de figuras

```java
int tamX, tamY;
void setup(){
    size(300,300);
    tamX=0; tamY=0;
    ellipseMode(CENTER);
}
void draw(){
    background(200);
    ellipse(0,0,tamX, tamY);
    tamX++;
    tamY++;
}
```

### Colores

```java
int tamX, tamY;
void setup(){
    size(300,300);
    tamX=0; tamY=0;
    colorMode(width);
}
void draw(){
    background(200);
    fill(tamX);
    rect(0,0,tamX, 50);
    fill(tamX, 0, 0);
    rect(0,50,tamX, 50);
    fill(0, tamX, 0);
    rect(0,100,tamX, 50);
    fill(0, 0, tamX);
    rect(0,150,tamX, 50);
    tamX++;
    tamY++;
}
```

### Div (/) vs Mod (%)

```java
int inc, tam;
void setup(){
    size(500,400);
    textSize(30);
    inc = 0; tam = 50;
    frameRate(10);
}
void draw(){
    background(200);
    text("inc: " + inc, 50, 25);
    rect(inc, 30, 40, 40);
    
    text("inc / tam: " + (inc / tam), 50, 100);
    rect(inc / tam, 100, 40, 40);
    
    text("floor(inc / tam) * tam: " + (floor(inc / tam) * tam), 50, 175);
    rect((floor(inc / tam) * tam), 175, 40, 40);
    
    text("inc % tam: " + (inc % tam), 50, 250);
    rect(inc % tam, 250, 40, 40);
    
    inc++;
}
```

### Recuadros aleatorios

```java
int posX, posY;
int cant, tamX, tamY;
void setup(){
    size(500,300);
    cant=10;
    tamX=width/cant;
    tamY=height/cant;
}
void draw(){
    fill(random(0,255),random(0,255),random(0,255));
    posX = round(random(0,width) / tamX) * tamX;
    posY = round(random(0,height) / tamY) * tamY;
    rect(posX, posY, tamX, tamY);
}
```

### Texto en movimiento

```java
int posY;
float fontTam;
void setup(){
    size(600, 300);
    fontTam = 40;
    textSize(fontTam);
    posY=0;
}
void draw(){
    background(200);
    textSize(fontTam);
    text("Bienvenidos!", 100, height - posY);
    posY++;
    fontTam = fontTam - 0.5;
}
```
