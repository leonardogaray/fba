## Ejercicio 8 - Paso 3
Limitamos las zonas verticales (en Y) donde se puede dibujar, haciendo uso del operado AND (&&). Tenemos que agregar las mismas a las condiciones puestas previamente. De esta manera se van a tener que dar las cuatro condiciones como verdaderas para que se pueda dibujar 

```java
void setup(){
    size(400,400);
    rectMode(CENTER);
}
void draw(){
    //Rectangulo que muestra la zona delimitada sin relleno
    noFill();
    rect(width/2, height/2, 200, 200);
    //Si la posicion en X del mouse es mayor a 100 Y la posicion en X del mouse es menor a 300
    // Y si la posicion en Y del mouse es mayor a 100 Y la posicion en Y del mouse es menor a 300
    if(mouseX > 100 && mouseX < 300 && mouseY > 100 && mouseY < 300){
        //Dibujamos como si fuese con un lapiz
        line(mouseX, mouseY, pmouseX, pmouseY);
    }
}
```

## Ejercicio 8 - Paso 4
Podemos notar que si movemos el mouse demasiado rápido, hay líneas que se encuentran por fuera de la zona delimitada, lo que se produce porque se cumple la condición para mouseX y mouseY, pero no pmouseX y pmouseY, que se encuentran por fuera de la zona. Para poder resolver esto debemos agregar a nuestra condición, que se tomen en cuenta los valores de pmouseX y pmouseY.

```java
void setup(){
    size(400,400);
    rectMode(CENTER);
}
void draw(){
    //Rectangulo que muestra la zona delimitada sin relleno
    noFill();
    rect(width/2, height/2, 200, 200);
    //Si la posicion en X del mouse es mayor a 100 Y la posicion en X del mouse es menor a 300
    // Y si la posicion en Y del mouse es mayor a 100 Y la posicion en Y del mouse es menor a 300
    if(mouseX > 100 && mouseX < 300 && mouseY > 100 && mouseY < 300){
        if(pmouseX > 100 && pmouseX < 300 && pmouseY > 100 && pmouseY < 300){
            //Dibujamos como si fuese con un lapiz
            line(mouseX, mouseY, pmouseX, pmouseY);
        }
    }
}
```

## Ejercicio 9 - Paso 1
Dibujamos las 2 elipses sin relleno para indicar la zona sobre la cual debemos dibujar, y aprovechamos para dibujar como si fuese con un lapiz con el mouse

```java
void setup(){
    size(400,400);
}
void draw(){
    //Elipses que muestran la zona delimitada sin relleno
    noFill();
    ellipse(width/2, height/2, 200, 200);
    ellipse(width/2, height/2, 100, 100);
    //Dibujamos como si fuese con un lapiz
    line(mouseX, mouseY, pmouseX, pmouseY);
}
```

## Ejercicio 9 - Paso 2
A diferencia de las zonas cuadradas, para poder delimitar las zonas circulares debemos hacerlo con la función dist, tomando como base la distancia que existe en el centro de la pantalla y el radio de la elipse

```java
void setup(){
    size(400,400);
}
void draw(){
    //Elipses que muestran la zona delimitada sin relleno
    noFill();
    ellipse(width/2, height/2, 200, 200);
    ellipse(width/2, height/2, 100, 100);
    //Si la distancia que hay entre el centro de la pantalla y la posicion del mouse es menor al radio, 
    //entonces permitimos que se pueda dibujar
    if(dist(width/2, height/2, mouseX, mouseY)<100){
        //Dibujamos como si fuese con un lapiz
        line(mouseX, mouseY, pmouseX, pmouseY);
    }
}
```

## Ejercicio 9 - Paso 3
En el paso anterior podemos dibujar cuando la distancia que hay entre el mouse y el centro es menor al radio, pero, ahora necesitamos que no se dibuje dentro de la elipse mas chica, por lo que la condición ahora debe ser la opuesta, o sea, que la distancia que hay entre el centro de la pantalla y el mouse sea mayor al radio de la elipse mas pequeña.
En consecuencia, el radio debe ser menor a 100 y mayor a 50.

```java
void setup(){
    size(400,400);
}
void draw(){
    //Elipses que muestran la zona delimitada sin relleno
    noFill();
    ellipse(width/2, height/2, 200, 200);
    ellipse(width/2, height/2, 100, 100);
    //Si la distancia que hay entre el centro de la pantalla y la posicion del mouse es menor al radio 100, 
    //entonces permitimos que se pueda dibujar
    // Y
    //Si la distancia que hay entre el centro de la pantalla y la posicion del mouse es mayor al radio 50, 
    // entonces permitimos que se pueda dibujar
    if(dist(width/2, height/2, mouseX, mouseY)<100 && dist(width/2, height/2, mouseX, mouseY)>50){
        //Dibujamos como si fuese con un lapiz
        line(mouseX, mouseY, pmouseX, pmouseY);
    }
}
```

## Ejercicio 9 - Paso 4
Ahora, el problema con pmouseX y pmouseY sigue presente, por lo que podemos agregar la condición que nos permita evitar ese problema.

```java
void setup(){
    size(400,400);
}
void draw(){
    //Elipses que muestran la zona delimitada sin relleno
    noFill();
    ellipse(width/2, height/2, 200, 200);
    ellipse(width/2, height/2, 100, 100);
    //Si la distancia que hay entre el centro de la pantalla y la posicion del mouse es menor al radio 100, 
    //entonces permitimos que se pueda dibujar
    // Y
    //Si la distancia que hay entre el centro de la pantalla y la posicion del mouse es mayor al radio 50, 
    // entonces permitimos que se pueda dibujar
    if(dist(width/2, height/2, mouseX, mouseY)<100 && dist(width/2, height/2, mouseX, mouseY)>50){
        if(dist(width/2, height/2, pmouseX, pmouseY)<100 && dist(width/2, height/2, pmouseX, pmouseY)>50){
            //Dibujamos como si fuese con un lapiz
            line(mouseX, mouseY, pmouseX, pmouseY);
        }
    }
}
```

```java
void setup(){
    size(400,400);
    //Rectangulo que muestra la zona delimitada sin borde
    fill(150);
    noStroke();
    rect(50, 0, 50, height);
    rect(150, 0, 50, height);
    rect(250, 0, 50, height);
}
void draw(){
    //Si la posicion en X del mouse es mayor a 50 Y la posicion en X del mouse es menor a 100 
    // O
    //Si la posicion en X del mouse es mayor a 150 Y la posicion en X del mouse es menor a 200
    if( (mouseX > 50 && mouseX < 100) || (mouseX > 150 && mouseX < 200) ){
        //Dibujamos como si fuese con un lapiz
        stroke(0);
        line(mouseX, mouseY, pmouseX, pmouseY);
    }
}
```

## Ejercicio 16
Armar un cañon que dispara al hacer click y se mueve con la posición en Y del mouse.

## Ejercicio 16 - Paso 1
Hacemos el cañón. Para realizar el mismo solo lo vamos a representar con un rectángulo que acompaña la posición en y del mouse (mouseY)

```java
void setup(){
    size(400,400);
}
void draw(){
    background(200);
    rect(0, mouseY, 100, 50);
}
```

## Ejercicio 16 - Paso 2
EL cañón es la parte fácil, ahora debemos dibujar la bala. No nos preocupemos en las partes relacionadas a como dispararla, o como debe seguir al cañón, sino en tan solo hacer la animación de la bala moviendose desde la izquierda hacia la derecha.

```java
int posXBala;
void setup(){
    size(400,400);
    posXBala = 0;
}
void draw(){
    background(200);
    ellipse(posXBala,100,25,25);
    rect(0, mouseY, 100, 50);
    posXBala++;
}
```

## Ejercicio 16 - Paso 3
Ahora, vamos a hacer que solo se ejecute la animación si hice click con el mouse. Vamos a tomar una variable booleana que me permita guardar el valor de si se hizo click, y de esa manera permitir que la variable posXBala se incremente. Esta variable booleana va a cambiar cuando se ejecute el evento mouseClicked.

```java
int posXBala;
boolean balaDisparada;
void setup(){
    size(400,400);
    posXBala = 0;
    balaDisparada = false;
}
void draw(){
    background(200);
    ellipse(posXBala,100,25,25);
    rect(0, mouseY, 100, 50);
    //En caso de que la variable balaDisparada sea igual a TRUE, solo ahi se incrementara el valor.
    if(balaDisparada){
        posXBala++;
    }
}
void mouseClicked(){
    balaDisparada = true;
}
```

## Ejercicio 16 - Paso 4
La bala no debería de salir desde cualquier lado, sino desde la punta del cañon, por lo cual, ahora debemos modificar su posición en Y de acuerdo a la posición del mismo. Creamos una nueva variable posYBala que vaya cambiando su valor de acuerdo a la posición en y del cañón (mouseY).

```java
int posXBala, posYBala;
boolean balaDisparada;
void setup(){
    size(400,400);
    posXBala = 0;
    balaDisparada = false;
}
void draw(){
    background(200);
    //Le sumamos 75 a posXBala para poder ubicarla al final del canon y que la misma no se vea
    //Le sumamos 25 a posYBala para poder ubicarla al centro del canon
    ellipse(posXBala + 75,posYBala + 25,25,25);
    rect(0, mouseY, 100, 50);
    //En caso de que la variable balaDisparada sea igual a TRUE, solo ahi se incrementara el valor.
    if(balaDisparada){
        posXBala++;
    }
    posYBala = mouseY;
}
void mouseClicked(){
    balaDisparada = true;
}
```

## Ejercicio 16 - Paso 5
Al disparar la bala, la misma sigue la posición del mouse en Y, lo cual no es correcto. Debemos buscar la manera de que la bala cambie su ubicación solo cuando la bala no ha sido disparada, por lo que haremos uso del else.

```java
int posXBala, posYBala;
boolean balaDisparada;
void setup(){
    size(400,400);
    posXBala = 0;
    balaDisparada = false;
}
void draw(){
    background(200);
    //Le sumamos 75 a posXBala para poder ubicarla al final del canon y que la misma no se vea
    //Le sumamos 25 a posYBala para poder ubicarla al centro del canon
    ellipse(posXBala + 75,posYBala + 25,25,25);
    rect(0, mouseY, 100, 50);
    //En caso de que la variable balaDisparada sea igual a TRUE, solo ahi se incrementara el valor.
    if(balaDisparada){
        posXBala++;
    }else{
        posYBala = mouseY;
    }
}
void mouseClicked(){
    balaDisparada = true;
}
```

## Ejercicio 16 - Paso 6
Como último paso, solo dibujamos la bala cuando ha sido disparada, y aumentamos su velocidad.

```java
int posXBala, posYBala;
boolean balaDisparada;
void setup(){
    size(400,400);
    posXBala = 0;
    balaDisparada = false;
}
void draw(){
    background(200);
    rect(0, mouseY, 100, 50);
    //En caso de que la variable balaDisparada sea igual a TRUE, solo ahi se incrementara el valor.
    if(balaDisparada){
        //Le sumamos 75 a posXBala para poder ubicarla al final del canon y que la misma no se vea
        //Le sumamos 25 a posYBala para poder ubicarla al centro del canon
        ellipse(posXBala + 75,posYBala + 25,25,25);
        //Aumentamos 10 veces la velocidad
        posXBala+=10;
    }else{
        posYBala = mouseY;
    }
}
void mouseClicked(){
    balaDisparada = true;
}
```
