# Condicionales


```java
PImage parallax1, parallax2, parallax3, parallax4;
Float posX0, posX1, posX2, posX3, posX4;
Integer dir = 1;

void setup(){
    size(400, 300);
    parallax1 = loadImage("clase6/parallax-1.png");
    parallax2 = loadImage("clase6/parallax-2.png");
    parallax3 = loadImage("clase6/parallax-3.png");
    parallax4 = loadImage("clase6/parallax-4.png");
}
void draw(){
    background(200);
    image(parallax1, posX1, 0);
    image(parallax2, posX2, 20);
    image(parallax3, posX3, 40);
    image(parallax4, posX4, 230);
    
    if(keyPressed){
        if(keyCode == RIGHT){
            dir = -1;
        }
        if(keyCode == LEFT){
            dir = 1;
        }
    
        posX3 = posX3 + 5 * dir;
        posX2 = posX2 + 3 * dir;
        posX1 = posX1 + 1 * dir;
        posX4 = posX4 + 10 * dir;
    }
}
```
